```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('main.db')
```

```python
authors_df = pd.read_csv('authors.csv')
authors_df.to_sql('authors_raw', conn, if_exists = 'replace')
display(authors_df)

papers_df = pd.read_csv('papers.csv')
papers_df.to_sql('papers_raw', conn, if_exists = 'replace')
display(papers_df)
```
